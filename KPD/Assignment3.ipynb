# Mean and Variance
A program to compute the mean and variance of 1000 data

```python
import random

def gen_random(n):
  dataset = list()
  for i in range(n):
    rand = random.randint(1,100)
    dataset.append(rand)
  return dataset

def count_mean(x):
  sum = 0
  count = len(x)
  for i in range(count):
    sum = sum+x[i]
  return sum/count

def variance(x):
  sum = 0
  mean = count_mean(x)
  count = len(x)
  for i in range(count):
    sum = sum + (x[i]-mean)**2
  return sum/count


dataset = gen_random(1000)
print("Mean : ", count_mean(dataset))
print("Variance : ", variance(dataset))
```

# Recursive 2^(n-1)
Recursive function for An = 2^(n-1) with A1 = 1

```python
n = 10

def sequence(n):
  if(n<=1):
    return 1
  else:
    seq = 2**(n-1)
    return sequence(n-1), seq

sequence(n)
```

# Function t
Given function f(t), calculate each step for the functions

```python
import math
import matplotlib.pyplot as plt
import pandas as pd

delta = 0.25
init_val = 0
final_val = 10
def function(a,b,delta):
  iteration = math.ceil((b-a)/delta)+1
  x = list()
  y = list()
  for i in range(iteration):
    t = delta*i
    x.append(t)
    y.append((t**(1/2)*(t-5)**2)/4)
  return x,y

x,y = function(init_val,final_val,delta)
df = pd.DataFrame({"t" : x, 
                   "f(t)" : y})
df
```

```python
plt.plot(x, y)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Function graph')
plt.figure(figsize=(9, 6))
plt.show()
```
